```python
def INV(data , cap , n):
    import pandas as pd
    
    month = ["Jan","Feb","Mar","Apr","May","Jun","July","Aug","Sep","Out","Nov","Dec"]
        
    capacity = []                                            
    for i in range (0,12):

        capacity.append(cap)
        
    extra_quantity = []                                                                                      
    
    Remaining_quantity = []                                    
    
    adj_quantity = []                                        
    for i in range (0,12):

        j = data[i] - cap                           
        if j <= 0 :
            j = 0
        extra_quantity.append(j)

        k = cap - data[i]
        if k <= 0 :
            k = 0
        Remaining_quantity.append(k)
        
        l = data[i]
        if l > cap :
            l = cap 
        adj_quantity.append(l)
    
        
    new_Remaining_quantity = Remaining_quantity
    new_extra_quantity = extra_quantity
    
    
    extra_quantity_month = []
    Remaining_quantity_month = []
    fullfill_quantity = []
    
    
    for i in range(len(extra_quantity)):
        while extra_quantity[i] > 0:
                for j in range(1, n + 1):
                    k = i - j
                    if k >= 0:

                        if Remaining_quantity[i - j] == 0:                            
                            pass
                        else:
                            diff = extra_quantity[i] - Remaining_quantity[i - j]
                            
                            
                            
                            mm = month[i]
                            extra_quantity_month.append(mm)
                            
                            nn = month[i - j]
                            Remaining_quantity_month.append(nn)                            
                            dif =  extra_quantity[i] - abs(diff)
                            
                            fullfill_quantity.append(abs(dif))
                            if extra_quantity[i] < Remaining_quantity[i - j]:
                                fullfill_quantity[-1]=extra_quantity[i]
                                
                           
                        
                            if diff < 0:
                                Remaining_quantity[i - j] = abs(diff)
                                extra_quantity[i] = 0
                            else:
                                Remaining_quantity[i - j] = 0
                                extra_quantity[i] = abs(diff)
           
                           
                i = i+1
                if i > 11 :
                    break

    
    
    for i in range (0,12):
        n = Remaining_quantity[i]
        if n == 0 :
            adj_quantity[i] = cap
        if n > 0 :
            m = cap - Remaining_quantity[i]
            adj_quantity[i] = m
            
    
            
        
    MONTH = month
    DATA = data
    EQ = new_extra_quantity
    RQ = new_Remaining_quantity
    d = {'MONTH': MONTH,'CAPACITY':capacity, 'DEMAND': DATA, 'adj_quntity':adj_quantity, 'Extra_Quntity': EQ, 'Remaining_Quntity': RQ  }
    df = pd.DataFrame(d)
     
    d2 = {'extra_quantity_month' : extra_quantity_month,  'Remaining_quantity_month' : Remaining_quantity_month ,  'fullfill_quantity' : fullfill_quantity }
    df2 = pd.DataFrame(d2)
    result = df,df2
    return df ,df2


```
